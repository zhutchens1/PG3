# Test velocity definitions

```python
import pandas as pd
import numpy as np
from astropy.cosmology import LambdaCDM
import matplotlib.pyplot as plt

cosmo = LambdaCDM(70., 0.3, 0.7)
```

```python
df = pd.read_csv("/srv/one/zhutchen/g3groupfinder/resolve_and_eco/ECOdata_G3catalog_luminosity.csv")
```

```python
coma = df[df.g3grp_l==14]
```

# Peculiar Velocities

```python
plt.figure()
bins = np.arange(-2000,1050,200)
plt.hist(coma.cz - coma.g3grpcz_l, bins=bins)
vpec = 3e5 * (coma.cz/3e5 - coma.g3grpcz_l/3e5)/(1+coma.g3grpcz_l/3e5)
plt.hist(vpec, bins=bins, histtype='step', linewidth=3)
plt.show()

plt.figure()
plt.hist(vpec - (coma.cz - coma.g3grpcz_l))
plt.show()
```
