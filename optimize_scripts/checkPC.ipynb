```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import mode
```

```python
f = '/srv/two/zhutchen/paper3/catalogs/realistic_lsdr9/eco_Pth0.01.csv'
```

```python
eco = pd.read_csv(f).set_index('name')
```

```python
eco
```

```python
def get_stats(df, g3grpid):
    g3 = df[df.g3grp_l==g3grpid]
    print('original galx: ', g3.index.to_numpy())
    print('pg3 IDs: ', g3.pg3grp.to_numpy())
    pg3mode = mode(g3.pg3grp.to_numpy())
    print('mode pg3_id = {a}'.format(a=pg3mode.mode))
    pg3grp = df[df.pg3grp==pg3mode.mode]
    print('####################################')
    print('G3 IDs in test group: ', pg3grp.g3grp_l.to_numpy())
    print('pg3 P_TR: ', pg3grp.P_TR.to_numpy())
    print('pg3 C_TR: ', pg3grp.C_TR.to_numpy())
    print('#####################################')
    print('pg3 P_RT: ', pg3grp.P_TR.to_numpy())
    print('pg3 C_RT: ', pg3grp.C_TR.to_numpy())
```

# G3 group 5 (N=4)

```python
get_stats(eco, 5)
```

# G3 group 42 (N=4)

```python
get_stats(eco, 42)
```
